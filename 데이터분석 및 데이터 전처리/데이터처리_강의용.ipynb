# 데이터 조작하기

## 1. np.where 방식

```python
customerData.columns
```

```python
customerData["PRODUCTAGE_NEW"] = np.where(customerData.PRODUCTAGE < 1, 1, 0)
```

```python
# 검증과정 / 로직을 짜고 반대 케이스로 검증하기
customerData.loc[(customerData.PRODUCTAGE < 1) &
                 (customerData.PRODUCTAGE_NEW !=1)]
```

```python
# 다중 조건 / 1보다 작을 땐 1, 2보다 작을 땐 2, 나머지는 3 / if, elif, else랑 비슷함
customerData["PRODUCTAGE_NEW"] = np.where(customerData.PRODUCTAGE < 1, 1, 
                                         np.where(customerData.PRODUCTAGE < 2, 2, 3))
customerData.head(5)
```

## 2. 함수 활용

```python
# 함수 정의 
# inputValue < 1 : 1 
# 1 <= inputValue < 2 : 2
# inputValue >= 2 : 3
```

```python
def valueClustering(inValue):
    # inValue = 0.3
    outValue = inValue

    if inValue < 1:
        outValue = 1
    elif inValue < 2:
        outValue = 2
    else: 
        outValue = 3
    return outValue
```

```python
valueClustering(1.2)
```

```python
# 함수 적용
customerData["PRODUCTAGE_NEW"] = customerData.PRODUCTAGE.apply(valueClustering)
```

```python
customerData.head(5)
```
