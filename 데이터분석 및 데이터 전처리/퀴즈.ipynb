```python
import pandas as pd
```

```python
selloutData = pd.read_csv("../stddev/dataset/kopo_channel_seasonality_new.csv", encoding="ms949")
```

```python

# 3. 타겟 컬럼의 데이터 타입을 str로 변경
selloutData["YEARWEEK"] = selloutData["YEARWEEK"].astype(str)

# 4. 인덱스를 통해 "YEAR"컬럼과 "WEEK"컬럼으로 분리
selloutData["YEAR"] = selloutData["YEARWEEK"].str[0:4]
selloutData["WEEK"] = selloutData["YEARWEEK"].str[-2:]

# 5. WEEK의 53주차를 제거하기 위한 조건을 적용하기 위해 int로 변경
selloutData["WEEK"] = selloutData["WEEK"].astype(int)

# 조건 적용 후 출력 
# refinedData = selloutData.loc[selloutData.REGIONID]
selloutData
```

```python
# 1번 문제
data1=selloutData.loc[(selloutData.YEAR == '2015') & (selloutData.REGIONID =='A01')]
groupKey = ["PRODUCT"]
groupData1 = data1.groupby(groupKey)["QTY"].agg("mean").reset_index()
pd.DataFrame(groupData1.sort_values(by='QTY', ascending=False))
```

```python
# 2번 문제
data2=selloutData.loc[(selloutData.YEAR == '2015') & (selloutData.REGIONID =='A01')]
groupKey = ["PRODUCT"]
groupData2 = data2.groupby(groupKey)["QTY"].agg("mean").reset_index()
pd.DataFrame(groupData2.sort_values(by='PRODUCT', ascending=True))
```

```python
# 3번 문제
data3=selloutData.loc[(selloutData.WEEK < 14) & (selloutData.YEAR == '2015')]
groupKey = ["REGIONID"]
groupData3 = data3.groupby(groupKey)["QTY"].agg("mean").reset_index()
pd.DataFrame(groupData3.sort_values(by='QTY', ascending=False))
```

```python
# 4번 문제
dataQ1=selloutData.loc[selloutData.WEEK < 14]
groupKey = ["REGIONID","PRODUCT"]
groupData4 = dataQ1.groupby(groupKey)["QTY"].agg("mean").reset_index()
pd.DataFrame(groupData4.sort_values(by='QTY', ascending=False))
```
